# Seaborn liibrary part 2

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df1 = sns.load_dataset("penguins")
```

```python
df1
```

```python
sns.barplot(x = df1.island,y = df1.bill_length_mm)
plt.show()
```

```python
sns.barplot(x = "island",y = "bill_length_mm",data = df1)
plt.show()
```

```python
sns.barplot(x = "island",y = "bill_length_mm",data = df1, hue = "sex") # HUe is used to categorize the system:
plt.show()
```

# Order

```python
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm",data = df1,hue = "sex",order = order1)
plt.show()
```

# hue order

```python
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm",data = df1,hue = "sex",order = order1,hue_order = ['Female','Male'])
plt.show()
```

# Change the colors

```python
sns.barplot(x = "island", y = "bill_length_mm",data = df1,hue = "sex",palette = "Accent")
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm",data = df1,hue = "sex",palette = "prism")
plt.show()
```

```python
sns.barplot(x = "island",y = "bill_length_mm",data = df1,hue = "sex",palette = "gist_earth")
plt.show()
```

```python
sns.barplot(x = "island",y = "bill_length_mm",data = df1,hue = "sex",hue_order = ['Female','Male'],palette = "gist_earth")
plt.show()
```

# Saturation

```python
sns.barplot(x = "island", y = "bill_length_mm",data = df1,hue = "sex",palette = "prism")
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm",data = df1,hue = "sex",palette = "prism",saturation = 0.5)
plt.show()
```
