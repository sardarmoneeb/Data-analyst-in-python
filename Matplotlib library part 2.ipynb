# Pie chart

```python
x = [20,40,60,80]
y = ["SE","CS","IT","AI"]
plt.title("Pie chart")
plt.pie(x, labels = y)
plt.legend()# Sow different colors with names
plt.show()
```

# Stem Graph

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.title("Stem Graph")
plt.stem(x,y)
plt.show()
```

# Bar Graph

```python
x = [20,40,60,80]
y = [10,30,50,70]
c = ["red","brown","yellow","hotpink"]
plt.bar(x,y,color=c, width = 2.5)
plt.show()
```
