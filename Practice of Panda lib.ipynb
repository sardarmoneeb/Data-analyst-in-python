```python
import pandas as pd

# Data
data = {
    'Name': ['Moneeb', 'Saran', 'Aneeb', 'Moseeb', 'Bilal', 'Noor', 'Qura', 'Sana', 'Fatima', 'Zainab'],
    'Roll no': [1, 2, 4, 5, 63, 8, 10, 40, 76, 89],
    'City': ['Kotli', 'pindi', 'Bloach', 'Islamabad', 'Rawalkot', 'pounch', 'pounch', 'pounch', 'pounch', 'pounch']
}

# Lists of male and female names
male_names = ['Moneeb', 'Saran', 'Aneeb', 'Moseeb', 'Bilal']
female_names = ['Noor', 'Qura', 'Sana', 'Fatima', 'Zainab']

# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# Function to determine gender based on name
def determine_gender(name):
    if name in male_names:
        return 'Male'
    elif name in female_names:
        return 'Female'
    else:
        return 'Unknown'

# Apply the function to the DataFrame
df['Gender'] = df['Name'].apply(determine_gender)

print(df)




```
