```python
import pandas as pd

def clean_hospital_data(df):
    # Define column groups
    id_cols = ["MRD NO."]
    date_cols = ["D.O.D"]
    numeric_cols = ["HB", "TLC", "PLATELETS", "GLUCOSE", "UREA", "CREATININE", "BNP", "EF"]
    categorical_cols = ["CHEST INFECTION"]

    # --- IDs as string ---
    for col in id_cols:
        if col in df.columns:
            df[col] = df[col].astype(str)

    # --- Dates as datetime ---
    for col in date_cols:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], errors="coerce")

    # --- Numeric columns ---
    for col in numeric_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")
            df[col] = df[col].fillna(df[col].mean())  # fill with mean

    # --- Categorical columns ---
    for col in categorical_cols:
        if col in df.columns:
            df[col] = df[col].fillna("Unknown").astype(str)

    return df

# -------------------------
# Usage
df_adm = pd.read_excel("D:/Kaggle datasets/HDHI Admission data.xlsx")
df_adm = clean_hospital_data(df_adm)

print("✅ Cleaned data types:\n", df_adm.dtypes)
print("\n🔎 Remaining null values:\n", df_adm.isnull().sum())
```

```python
def smart_convert(df):
    for col in df.columns:
        # Skip if already numeric or datetime
        if pd.api.types.is_numeric_dtype(df[col]) or pd.api.types.is_datetime64_any_dtype(df[col]):
            continue

        # Clean values (remove commas, spaces)
        temp = df[col].astype(str).str.replace(",", "", regex=False).str.strip()

        # Try convert to numeric
        numeric = pd.to_numeric(temp, errors="coerce")

        if numeric.notna().sum() > len(df) * 0.6:  # agar majority numeric hai
            # Check if all non-null values are integers
            if (numeric.dropna() % 1 == 0).all():
                df[col] = numeric.astype("Int64")   # safe integer
                print(f"🔢 {col} → Int64")
            else:
                df[col] = numeric.astype("float64")  # otherwise float
                print(f"🔢 {col} → float64")
            continue

        # Try convert to datetime
        try:
            converted = pd.to_datetime(df[col], errors="coerce", dayfirst=True)
            if converted.notna().sum() > len(df) * 0.5:
                df[col] = converted
                print(f"📅 {col} → datetime64")
                continue
        except:
            pass

        # Default: keep as string
        df[col] = df[col].astype(str)
        print(f"🔤 {col} → object")

    return df
     

df_adm = pd.read_excel("D:/Kaggle datasets/HDHI Admission data.xlsx")
df_adm = smart_convert(df_adm)

print("\n✅ Final column types:\n", df_adm.dtypes.to_string())
```

```python
cols_to_int = ["PM10 MAX", "NO2 MAX", "OZONE MAX"]

# Strings ko force karke numeric mein convert karna (errors='coerce' non-numeric ko NaN bana dega)
df_poll[cols_to_int] = df_poll[cols_to_int].apply(pd.to_numeric, errors="coerce")

# Ab NaN ko mean se fill karna
df_poll[cols_to_int] = df_poll[cols_to_int].fillna(df_poll[cols_to_int].mean())

# Ab int mein convert karna
df_poll[cols_to_int] = df_poll[cols_to_int].astype(int)
```
