```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [4,2,6,7,8]
plt.plot(x,y, marker = "o")
plt.show()
```

```python
x= [10,20,30,40,50]
y = [5,15,25,35,45]
plt.title("Bar Graph")
plt.xlabel("This is X axis")
plt.ylabel("This is Y axis")
c = ['green','yellow','red','blue','pink']
plt.bar(x,y,color = c, width = 2)
plt.show()
```

```python
x = [5,10,15,20]
y = ["CS","IT","SE","AI"]
plt.pie(x,labels=y)
plt.legend()
plt.show()
```

```python
import seaborn as sns
x = [1,2,3,4]
y = [2,4,6,8]
sns.lineplot(x=x,y=y)
plt.show()
```

```python
import seaborn as sns
x = [1,2,3,4]
y = [2,4,6,8]
plt.title("Barplot")
sns.barplot(x=x,y=y,width = 0.5)
plt.show()
```

```python
import plotly.express as px
```

```python
x = [1,2,3,4]
y = [2,4,6,8]
fig = px.line(x=x,y=y,markers = "o")
fig.show()
```

```python
import plotly.graph_objects as go
x1=[1,2,3,4,5]
y1=[2,3,5,7,8]
x2 = [5,6,7,9,14]
y2 = [8,6,9,12,15]
fig = go.Figure()
fig.update_layout(title= "Scatter Graph",xaxis_title="This is X axis",yaxis_title="This is Y axis")
fig.add_trace(go.Scatter(x=x1,y=y1,mode = "markers+lines",name="trace 1"))
fig.add_trace(go.Scatter(x=x2,y=y2,name = "trace 2"))
fig.show()
```
