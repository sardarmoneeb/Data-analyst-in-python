# Matplotlib library 

# Linear Graph

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
c = 'r'
plt.title("Linear Graph")
plt.plot(x,y,c)
plt.show()
```

```python
x = [2,4,8,10,15]
y = [12,4,10,8,7]

plt.plot(x,y,marker = "o")
plt.show()
```

```python
x = [2,4,8,10,15]
y = [12,4,10,8,7]

plt.plot(x,y)
plt.plot(x, marker = "o")
plt.plot(y, marker = "o")

plt.show()
```

# Bar chart

```python
x = [2,4,7,9,15]
y = [4,8,20,13,25]
c = ['r','g','b','hotpink','black']
plt.title("Bar Chart")
plt.bar(x,y,color = c, width = 0.8)
plt.show()
```

```python
import numpy as np
```

```python
x = np.array([10,15,20,25,30])
y = np.array([8,10,15,25,28])
plt.title("Bar Chart")
plt.barh(x,y, color = "hotpink",height = 1)
plt.show()
```

# Scatter Graph

```python
x = [1,4,8,12,16]
y = [2,4,10,25,20]
c = ["red","blue","brown","yellow","green"]
plt.title("Scatter Graph")
plt.xlabel("Month")
plt.ylabel("Date")
plt.scatter(x,y,color = c, s = 200)
plt.show()
```
