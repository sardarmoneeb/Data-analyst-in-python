```python
num_cols = df.select_dtypes(include='number').columns

for col in num_cols:
    Q1 = df[col].quantile(0.25)   # current column ka 25% value
    Q3 = df[col].quantile(0.75)   # current column ka 75% value
    IQR = Q3 - Q1

    df = df[(df[col] >= (Q1 - 1.5*IQR)) & (df[col] <= (Q3 + 1.5*IQR))]
```

```python

  import pandas as pd

# Outlier removal se pehle ka data (original dataset)
df_before = pd.read_excel("D:/Kaggle datasets/clean_online_retail.xlsx")

# Outlier removal ke baad ka data (cleaned dataset)
df_after = df_before[(df_before['UnitPrice'] >= 0) & (df_before['UnitPrice'] <= 10)]  # Example condition

# Sirf numeric columns select karein
num_cols = df_before.select_dtypes(include=['number']).columns

for col in num_cols:
    print(f"=== {col} ===")
    print("Before outlier removal:")
    print(df_before[col].describe())
    print("\nAfter outlier removal:")
    print(df_after[col].describe())
    print("\n" + "-"*60 + "\n")
  
```
