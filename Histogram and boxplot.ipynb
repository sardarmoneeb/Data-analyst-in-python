Histogram plot for Univarient Visualization

```python
(df["SepalLengthCm"].dtype)
```

```python

sns.histplot(x = "SepalLengthCm",data = df, bins = 12,kde = True)
plt.title("Distribution of sepal length")
plt.xlabel("SepalLengthCm")
plt.ylabel("count")
plt.show()
```

Boxplot for univarient visualization

```python
sns.boxplot(x = ["SepalLengthCm"])
plt.title("Boxplot of Sepal Length")
plt.xlabel("")
plt.show()
```

```python
sns.histplot(x = "SepalLengthCm",y = "SepalWidthCm",data = df,hue="Species")
```

```python
sns.histplot(x= "SepalWidthCm",data = df, bins = 10,kde = True)
plt.title("Distribution of SepalWidthCm ")
plt.xlabel("SepalWidthCm")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x = ["SepalWidthCm","SepalLengthCm"])
plt.title("Boxplot of Sepal width")
plt.xlabel("")
plt.show()
```
