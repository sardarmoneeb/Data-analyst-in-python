```python
df.drop_duplicates(inplace=True)
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_")
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df
```

```python
df.dtypes
```

```python
df["Gross_Sales"] = df["Gross_Sales"].str.replace(",", "", regex=False)
df["Gross_Sales"] = pd.to_numeric(df["Gross_Sales"], errors="coerce")

df["Discounts"] = df["Discounts"].str.replace(r"[^\d.]", "", regex=True)
df["Discounts"] = pd.to_numeric(df["Discounts"], errors="coerce")
```

```python
import pandas as pd

# Example: clean specific columns
cols_to_clean = ["Sales", "COGS", "Profit"]   # replace with your columns

for col in cols_to_clean:
    df[col] = (
        df[col]
        .astype(str)                           # ensure all values are strings
        .str.replace(r"[^\d.-]", "", regex=True)  # remove everything except digits, minus, and decimal
    )
    df[col] = pd.to_numeric(df[col], errors="coerce")  # convert to numeric
```
